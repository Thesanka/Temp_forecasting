```python
import pandas as pd
import numpy as np
import tensorflow as tf
import os
import opendatasets as od
```

```python
od.download("https://www.kaggle.com/datasets/gayathrirprog/jena-climate-from-2009-to-2016")
```

```python
dataset = pd.read_csv("/content/jena-climate-from-2009-to-2016/jena_climate_2009_2016.csv")
```

```python
dataset
```

```python
dataset= dataset[5::6]
dataset
```

```python
dataset.index = pd.to_datetime(dataset['Date Time'],format="%d.%m.%Y %H:%M:%S")
dataset[:26]
```

```python
temp = dataset["T (degC)"]
temp.plot()
```

```python
def dataset_to_x_y (dataset,window_size):
  dataset_as_np = dataset.to_numpy()
  x=[]
  y=[]
  for i in range(len(dataset_as_np)-window_size):
    row=[[a] for a in dataset_as_np[i:i+5]]
    x.append(row)
    label = dataset_as_np[i+5]
    y.append(label)
  return np.array(x),np.array(y)


```

```python
WINDOW_SIZE=5
x,y = dataset_to_x_y(temp,WINDOW_SIZE)
x.shape,y.shape
```

```python
x_train,y_train = x[:60000], y[:60000]
x_val, y_val    = x[60000:65000], y[60000:65000]
x_test,y_test = x[65000:], y[65000:]
```

```python
x
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential([
    InputLayer((5,1)),
    LSTM(64),
    Dense(8,"relu"),
    Dense(1,"linear")
])

model.summary()
```

```python
cp = ModelCheckpoint("model/", save_best_only = True)
model.compile(loss=MeanSquaredError(),
              optimizer = Adam(learning_rate = 0.01),
              metrics=[RootMeanSquaredError()])
```

```python
model.fit(x_train,y_train,
          validation_data=(x_val,y_val),
          epochs = 10,
          callbacks=[cp])
```

```python
from tensorflow.keras.models import load_model
```

```python
model = load_model('model/')
```

```python
train_prdictions = model.predict(x_train).flatten()
train_results = pd.DataFrame(data={"train_prdictions": train_prdictions, "Actuals":y_train})
train_results
```

```python
import matplotlib.pyplot as plt
plt.plot(train_results["train_prdictions"][:100])
plt.plot(train_results["Actuals"][:100])
```

```python
val_prdictions = model.predict(x_val).flatten()
val_results = pd.DataFrame(data={"val_prdictions": val_prdictions, "Actuals":y_val})
val_results
```

```python
plt.plot(val_results["val_prdictions"][:100])
plt.plot(val_results["Actuals"][:100])
```

```python
test_prdictions = model.predict(x_test).flatten()
test_results = pd.DataFrame(data={"test_prdictions": test_prdictions, "Actuals":y_test})
test_results
```

```python
plt.plot(test_results["test_prdictions"][:100])
plt.plot(test_results["Actuals"][:100])
plt.legend(['test_prdictions',"Actuals"])
```
